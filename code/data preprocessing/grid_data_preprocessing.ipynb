```python
# Dealing with negatives
grid_data['sun_is_up'] = grid_data['ghi_w_per_m2'] > 5

night_mask = (grid_data['sun_is_up'] == False) & (grid_data['solar_energy_generation_mw'] < 0)
day_mask = (grid_data['sun_is_up'] == True) & (grid_data['solar_energy_generation_mw'] < 0)

grid_data.loc[night_mask, 'solar_energy_generation_mw'] = 0
grid_data.loc[day_mask, 'solar_energy_generation_mw'] = grid_data.loc[day_mask, 'solar_energy_generation_mw'].abs()

grid_data['solar_energy_generation_mw'] = grid_data['solar_energy_generation_mw'].clip(lower=0)

print(f"Fixed {night_mask.sum()} night-time rows to zero.")
print(f"Fixed {day_mask.sum()} day-time rows using modulus.")
```

```python
# Dealing with wind negatives
grid_data['wind_is_blowing'] = grid_data['wind_speed_ms'] > 2.5

wind_noise_mask = (grid_data['wind_is_blowing'] == False) & (grid_data['wind_energy_generation_mw'] < 0)
wind_sign_error_mask = (grid_data['wind_is_blowing'] == True) & (grid_data['wind_energy_generation_mw'] < 0)

grid_data.loc[wind_noise_mask, 'wind_energy_generation_mw'] = 0
grid_data.loc[wind_sign_error_mask, 'wind_energy_generation_mw'] = grid_data.loc[wind_sign_error_mask, 'wind_energy_generation_mw'].abs()

grid_data['wind_energy_generation_mw'] = grid_data['wind_energy_generation_mw'].clip(lower=0)

print(f"Fixed {wind_noise_mask.sum()} low-wind noise rows.")
print(f"Fixed {wind_sign_error_mask.sum()} wind sign-error rows.")
```

```python
# Resampling to Hourly

if grid_data.index.name != 'timestamp':
    grid_data = grid_data.set_index('timestamp')

# Create the complete hourly dataset
grid_hourly = grid_data.resample('H').agg({
    'date': 'first',
    'clock_time': 'first',
    'season': 'first',      
    'day_of_week': 'first',
    'temp_c': 'mean',
    'humidity_pct': 'mean',
    'wind_speed_ms': 'mean',
    'ghi_w_per_m2': 'mean',
    'dni_w_per_m2': 'mean',
    'dhi_w_per_m2': 'mean',
    'solar_energy_generation_mw': 'mean',
    'wind_energy_generation_mw': 'mean',
    'electric_demand_mw': ['mean', 'max', 'std'] # Keep the 'spikiness' metrics
}).copy()

grid_hourly.columns = [
    'date', 'clock_time', 'season', 'day_of_week', 'temp_c', 'humidity_pct', 'wind_speed_ms', 
    'ghi_wm2', 'dni_wm2', 'dhi_wm2', 'solar_gen_mw', 'wind_gen_mw', 
    'demand_avg_mw', 'demand_max_mw', 'demand_std_mw'
]

# Calculate Net Load
grid_hourly['net_load_mw'] = grid_hourly['demand_avg_mw'] - grid_hourly['solar_gen_mw'] - grid_hourly['wind_gen_mw']
```

```python
# Check for any hour that didn't have 5-minute data
missing_hours = grid_hourly['demand_avg_mw'].isna().sum()
print(f"Number of empty hours in grid data: {missing_hours}")

if missing_hours > 0:
    grid_hourly = grid_hourly.interpolate(method='linear')
    print("Gaps filled using linear interpolation.")
```

```python
# Domain Specific Feature Engineering
import numpy as np

# 1. Cyclical Time Encoding (Crucial for LSTMs)
# This tells the model that Hour 23 is next to Hour 0
grid_hourly['hour_sin'] = np.sin(2 * np.pi * grid_hourly.index.hour / 24)
grid_hourly['hour_cos'] = np.cos(2 * np.pi * grid_hourly.index.hour / 24)

# 2. Momentum (Rolling Changes)
# Is demand ramping up? Positive = Increasing Stress
grid_hourly['demand_ramp'] = grid_hourly['demand_avg_mw'].diff()

# 3. Lag Features (The 'What happened before' context)
# Models love knowing the state of the grid 1 hour and 24 hours ago
grid_hourly['demand_lag_1h'] = grid_hourly['demand_avg_mw'].shift(1)
grid_hourly['net_load_lag_24h'] = grid_hourly['net_load_mw'].shift(24)

# 4. Rolling Averages (Smoothing the noise)
# 3-hour moving average for temperature (to see if buildings are getting heat-soaked)
grid_hourly['temp_rolling_3h'] = grid_hourly['temp_c'].rolling(window=3).mean()

# Since 'diff', 'shift', and 'rolling' create a few NaNs at the very top of the data...
grid_hourly = grid_hourly.fillna(method='bfill')
```

```python
# 0 = Weekend, 1 = Workday
grid_hourly['is_workday'] = grid_hourly['day_of_week'].isin(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']).astype(int)
```

```python
# Ratio of actual GHI to a 'perfect' sunny day (approximate max GHI is 1000)
grid_hourly['sky_clearness'] = grid_hourly['ghi_wm2'] / 1000
```

```python
# Volatility of demand over the last 6 hours
grid_hourly['demand_volatility_6h'] = grid_hourly['demand_avg_mw'].rolling(window=6).std()
```

```python
# 1. Solar Penetration 
grid_hourly['solar_penetration'] = grid_hourly['solar_gen_mw'] / (grid_hourly['demand_avg_mw'] + 1e-5)

# 2. Wind Penetration
grid_hourly['wind_penetration'] = grid_hourly['wind_gen_mw'] / (grid_hourly['demand_avg_mw'] + 1e-5)

# 3. Total Renewable Penetration (Combined)
# This is the Safety Buffer the grid has at any moment
grid_hourly['total_renewable_penetration'] = (grid_hourly['solar_gen_mw'] + grid_hourly['wind_gen_mw']) / (grid_hourly['demand_avg_mw'] + 1e-5)
```

```python
# Official NOAA (National Oceanic and Atmospheric Administration) Heat Index Equation.
def calculate_advanced_heat_index(temp_c, humidity):
    T = (temp_c * 9/5) + 32
    R = humidity
    
    # NOAA Regression Constants
    c1, c2, c3 = -42.379, 2.04901523, 10.14333127
    c4, c5, c6 = -0.22475541, -0.00683783, -0.05481717
    c7, c8, c9 = 0.00122874, 0.00085282, -0.00000199
    
    # 3. The Full Power Series
    hi_f = (c1 + (c2 * T) + (c3 * R) + (c4 * T * R) + (c5 * T**2) + 
            (c6 * R**2) + (c7 * T**2 * R) + (c8 * T * R**2) + (c9 * T**2 * R**2))
    
    return (hi_f - 32) * 5/9

grid_hourly['heat_index_c'] = calculate_advanced_heat_index(grid_hourly['temp_c'], grid_hourly['humidity_pct'])
```

```python
# Demand & Net Load Lags 
grid_hourly['demand_lag_24h'] = grid_hourly['demand_avg_mw'].shift(24)
grid_hourly['demand_lag_168h'] = grid_hourly['demand_avg_mw'].shift(168) # 1 Week
grid_hourly['net_load_lag_1h'] = grid_hourly['net_load_mw'].shift(1)

# Temperature Lags - For Heat Waves
grid_hourly['temp_lag_24h'] = grid_hourly['temp_c'].shift(24)
grid_hourly['temp_lag_48h'] = grid_hourly['temp_c'].shift(48)

# Renewable Lags
grid_hourly['solar_gen_lag_24h'] = grid_hourly['solar_gen_mw'].shift(24)
grid_hourly['wind_gen_lag_24h'] = grid_hourly['wind_gen_mw'].shift(24)

# Because a 168h shift creates 168 NaN rows at the very start...
grid_hourly = grid_hourly.fillna(method='bfill')
```

```python
# 1. Rolling Features 
grid_hourly['demand_rolling_mean_7d'] = grid_hourly['demand_avg_mw'].rolling(window=168).mean()
grid_hourly['demand_rolling_std_7d']  = grid_hourly['demand_avg_mw'].rolling(window=168).std()
grid_hourly['demand_rolling_max_24h'] = grid_hourly['demand_avg_mw'].rolling(window=24).max()

grid_hourly['temp_rolling_mean_7d']   = grid_hourly['temp_c'].rolling(window=168).mean()
grid_hourly['temp_rolling_max_3d']    = grid_hourly['temp_c'].rolling(window=72).max()

grid_hourly['solar_rolling_mean_7d']  = grid_hourly['solar_gen_mw'].rolling(window=168).mean()
grid_hourly['wind_rolling_std_24h']   = grid_hourly['wind_gen_mw'].rolling(window=24).std()

# 2. Degree Days (Physics of HVAC load)
# 18.3°C is the standard Balance Point for buildings
grid_hourly['cooling_degree_hours'] = (grid_hourly['temp_c'] - 18.3).clip(lower=0)
grid_hourly['heating_degree_hours'] = (18.3 - grid_hourly['temp_c']).clip(lower=0)

# 3. Accumulated Heat (The Heatwave Effect)
# This explains why the grid fails on the 3rd day of a heatwave even if temp is the same
grid_hourly['cooling_dh_cumsum_7d'] = grid_hourly['cooling_degree_hours'].rolling(window=168).sum()

# 4. Percentile Ranks 
# pct=True gives a 0-1 scale. 
grid_hourly['demand_percentile_rank'] = grid_hourly['demand_avg_mw'].rank(pct=True)
grid_hourly['temp_percentile_rank']   = grid_hourly['temp_c'].rank(pct=True)

# 5. Extreme Indicators 
demand_95th = grid_hourly['demand_avg_mw'].quantile(0.95)
temp_90th   = grid_hourly['temp_c'].quantile(0.90)

grid_hourly['is_extreme_demand'] = (grid_hourly['demand_avg_mw'] > demand_95th).astype(int)
grid_hourly['is_extreme_temp']   = (grid_hourly['temp_c'] > temp_90th).astype(int)

grid_hourly = grid_hourly.fillna(method='bfill')
```

```python
# Low Renewable + High Demand - Dark Doldrums Indicator
demand_median = grid_hourly['demand_avg_mw'].median()
grid_hourly['low_ren_high_dem'] = ((grid_hourly['total_renewable_penetration'] < 0.2) & 
                                   (grid_hourly['demand_avg_mw'] > demand_median)).astype(int)
```

```python
# 1. Momentum / Ramps
grid_hourly['net_load_ramp'] = grid_hourly['net_load_mw'].diff()
grid_hourly['solar_ramp'] = grid_hourly['solar_gen_mw'].diff()

# 2. Advanced Cyclical Time (Month and Day of Year)
grid_hourly['month'] = grid_hourly.index.month
grid_hourly['month_sin'] = np.sin(2 * np.pi * grid_hourly['month']/12)
grid_hourly['month_cos'] = np.cos(2 * np.pi * grid_hourly['month']/12)

grid_hourly['doy'] = grid_hourly.index.dayofyear
grid_hourly['doy_sin'] = np.sin(2 * np.pi * grid_hourly['doy']/365)
grid_hourly['doy_cos'] = np.cos(2 * np.pi * grid_hourly['doy']/365)

# 3. Stress Accumulation - for rolling count of Heat Wave days
is_hot = (grid_hourly['temp_c'] > 32).astype(int)
grid_hourly['heatwave_index'] = is_hot.rolling(window=72).sum() # Sum of hot hours in last 3 days

# 4. Weekends
grid_hourly['is_weekend'] = (grid_hourly.index.dayofweek >= 5).astype(int)
grid_hourly['is_peak_hour'] = grid_hourly.index.hour.isin([17, 18, 19, 20]).astype(int)

grid_hourly = grid_hourly.fillna(method='bfill')
```

```python
import holidays

regional_holidays = holidays.US(years=[2019, 2020, 2021])
grid_hourly['is_holiday'] = grid_hourly.index.map(lambda x: 1 if x in regional_holidays else 0)
grid_hourly['holiday_name'] = grid_hourly.index.map(lambda x: regional_holidays.get(x))
grid_hourly['is_pre_holiday'] = grid_hourly['is_holiday'].shift(-24).fillna(0)
grid_hourly['is_post_holiday'] = grid_hourly['is_holiday'].shift(24).fillna(0)
```
